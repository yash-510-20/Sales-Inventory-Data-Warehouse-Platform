# All Internal DB DF to be processed

```python
from pyspark.sql.window import Window

# 1) ORDER_HEADER*****
# Step 1: Add row number for duplicate handling
window_spec = Window.partitionBy("ORDER_NUMBER").orderBy(col("ORDER_NUMBER"))

df_o_header = df_ORDER_HEADER.withColumn("rn", row_number().over(window_spec))
# Step 2: Validation conditions
valid_condition = (
    col("ORDER_NUMBER").isNotNull() &
    col("RETAILER_NAME").isNotNull() &
    col("ORDER_DATE").isNotNull()
)
# Step 3: Valid DF → only first occurrence
df_valid_Header = df_o_header.filter(valid_condition & (col("rn") == 1)).drop("rn")

# Step 4: Rejected DF → nulls, wrong format, invalid date, duplicates
df_reject_Header = df_o_header.filter((~valid_condition) | (col("rn") > 1)).drop("rn")

#---------------------------------------------------------------------------------------------------------------------------

# 2) Product*****


# Step 1: Add row number for duplicate handling
window_spec = Window.partitionBy("PRODUCT_NUMBER").orderBy(col("PRODUCT_NUMBER"))

df_produ = df_PRODUCT.withColumn("rn", row_number().over(window_spec))

# Step 2: Validation conditions
valid_condition = (
    col("PRODUCT_NUMBER").isNotNull() &
    col("INTRODUCTION_DATE").isNotNull() &
    col("DISCONTINUED_DATE").isNotNull() &
    col("INTRODUCTION_DATE").rlike(r"^\d{4}-\d{2}-\d{2}$") &
    col("DISCONTINUED_DATE").rlike(r"^\d{4}-\d{2}-\d{2}$") &
    to_date(col("INTRODUCTION_DATE"), "yyyy-MM-dd").isNotNull() &
    to_date(col("DISCONTINUED_DATE"), "yyyy-MM-dd").isNotNull()
)
# Step 3: Valid DF → only first occurrence
df_valid_product = df_produ.filter(valid_condition & (col("rn") == 1)).drop("rn")

# Step 4: Rejected DF → nulls, wrong format, invalid date, duplicates
df_reject_product = df_produ.filter((~valid_condition) | (col("rn") > 1)).drop("rn")

#---------------------------------------------------------------------------------------------------------------------------

# 3) Order Details****

# Step 1: Add row number for duplicate handling
window_spec = Window.partitionBy("ORDER_DETAIL_CODE").orderBy(col("ORDER_DETAIL_CODE"))

df_Details = df_ORDER_DETAILS.withColumn("rn", row_number().over(window_spec))

# Step 2: Validation conditions
valid_condition = (
    col("ORDER_DETAIL_CODE").isNotNull() &
    col("QUANTITY").isNotNull() & (col("QUANTITY")>0) &
    col("UNIT_COST").isNotNull() & (col("UNIT_COST")>0) &
    col("UNIT_PRICE").isNotNull() & (col("UNIT_PRICE")>0) &
    col("UNIT_SALE_PRICE").isNotNull() & (col("UNIT_SALE_PRICE")>0) &
    col("SHIP_DATE").rlike(r"^\d{4}-\d{2}-\d{2}$") &
    to_date(col("SHIP_DATE"), "yyyy-MM-dd").isNotNull()
)

# Step 3: Valid DF → only first occurrence
df_valid_Details = df_Details.filter(valid_condition & (col("rn") == 1)).drop("rn")

 # Rejected DF → nulls, wrong format, invalid date, duplicates
df_reject_Details = df_Details.filter((~valid_condition) | (col("rn") > 1)).drop("rn")

# Step 4: Checking if ORDER_NUMBER is present in Oreder_Header else Reject If not present in Order_Header

df_valid_Details_fk =df_valid_Details.join(df_valid_Header.select("ORDER_NUMBER"), on="ORDER_NUMBER", how="inner")


# Step 5: Rejected DF → nulls, wrong format, invalid date, duplicates

df_reject_Details_fk = df_valid_Details.join(df_valid_Header,on="ORDER_NUMBER",how="left_anti")
# Step 6: Final rejected DF (combine all rejects)
df_reject_Details_final = df_reject_Details.union(df_reject_Details_fk)


#---------------------------------------------------------------------------------------------------------------------------

# 4) Return Item****

# Step 1: Add row number for duplicate handling
window_spec = Window.partitionBy("RETURN_CODE").orderBy(col("RETURN_CODE"))

df_items = df_RETURNED_ITEM.withColumn("rn", row_number().over(window_spec))

# Step 2: Validation conditions
valid_condition = (
    col("RETURN_CODE").isNotNull()
)

# Step 3: Valid DF → only first occurrence
df_valid_Return = df_items.filter(valid_condition & (col("rn") == 1)).drop("rn")

 # Rejected DF → nulls, wrong format, invalid date, duplicates
df_reject_Return = df_items.filter((~valid_condition) | (col("rn") > 1)).drop("rn")

# Step 4: Checking if ORDER_DETAIL_CODE is present in Order_Detail else Reject If not present in Order_Detail

df_valid_Return_fk = df_valid_Return.join(df_valid_Details_fk.select("ORDER_DETAIL_CODE"), on="ORDER_DETAIL_CODE", how="inner")


# Step 5: Rejected DF → nulls, wrong format, invalid date, duplicates

df_reject_Return_fk = df_valid_Return.join(df_valid_Details_fk,on="ORDER_DETAIL_CODE",how="left_anti")
# Step 6: Final rejected DF (combine all rejects)
df_reject_Return_final = df_reject_Return.union(df_reject_Return_fk)

#---------------------------------------------------------------------------------------------------------------------------

# 5) Inventory Levels****

# Step 1: Valid DF → only first occurrence & Validation conditions
df_valid_Inventory = df_INVENTORY_LEVELS.filter(col("INVENTORY_YEAR").isNotNull())

 # Rejected DF → nulls, wrong format, invalid date, duplicates
df_reject_Inventory = df_INVENTORY_LEVELS.filter(col("INVENTORY_YEAR").isNull())

# Step 2: Checking if PRODUCT_NUMBER is present in Product else Reject if not present in Product

df_valid_Inventory = df_valid_Inventory.join(df_valid_product.select("PRODUCT_NUMBER"), on="PRODUCT_NUMBER", how="inner")


# Step 3: Rejected DF → nulls, wrong format, invalid date, duplicates

df_reject_Inventory_fk= df_valid_Inventory.join(df_valid_product, on="PRODUCT_NUMBER", how="left_anti")
# Step 4: Final rejected DF (combine all rejects)
df_reject_Inventory_final = df_reject_Inventory.union(df_reject_Inventory_fk)

#---------------------------------------------------------------------------------------------------------------------------

# 6) Branch****
# there is no change in branch 
```
